```python

import pandas as pd
```

```python
# Load datasets
countries_df = pd.read_csv("https://gist.githubusercontent.com/sriharsha729/24a2854f2b602b172e7704aa80bd7cbc/raw/32a68a4abe0df64f863c2e456d59b58046407f7a/gistfile1.txt")
edu_df = pd.read_csv("https://gist.githubusercontent.com/sriharsha729/058a761615a3f494e30871bdabf67f38/raw/8a1d6910d6eb11c7f415807ba1d2426df7f7a7a7/gistfile1.txt")
world_data_df = pd.read_csv("https://gist.githubusercontent.com/sriharsha729/403eb6ec7f2a442844f85582503d81d8/raw/5ab995e783bb3e080b5ff7314e574e71eaeabaab/gistfile1.txt")


```

```python
# Display all columns from the countries dataset
print("Columns in countries dataset:")
print(countries_df.columns.tolist())

# Display all columns from the education dataset
print("\nColumns in education dataset:")
print(edu_df.columns.tolist())

# Display all columns from the world data dataset
print("\nColumns in world data dataset:")
print(world_data_df.columns.tolist())
```

```python
# Get null values for the countries dataset
print("Null values in countries dataset:")
print(countries_df.isnull().sum())

# Get null values for the education dataset
print("\nNull values in education dataset:")
print(edu_df.isnull().sum())

# Get null values for the world data dataset
print("\nNull values in world data dataset:")
print(world_data_df.isnull().sum())
```

```python
# Required columns for each dataset
required_columns_countries = [
    'country',
    'GDP: Gross domestic product (million current US$)',
    'GDP per capita (current US$)',
    'Education: Government expenditure (% of GDP)',
    'Education: Primary gross enrol. ratio (f/m per 100 pop.)',
    'Education: Secondary gross enrol. ratio (f/m per 100 pop.)',
    'Education: Tertiary gross enrol. ratio (f/m per 100 pop.)',
    'Unemployment (% of labour force)',
    'Population density (per km2, 2017)',
    'Life expectancy at birth (females/males, years)',
    'Urban population (% of total population)'
]

required_columns_education = [
    'Countries and areas',
    'Gross_Primary_Education_Enrollment',
    'Gross_Tertiary_Education_Enrollment',
    'Completion_Rate_Primary_Male',
    'Completion_Rate_Primary_Female',
    'Completion_Rate_Lower_Secondary_Male',
    'Completion_Rate_Lower_Secondary_Female',
    'Completion_Rate_Upper_Secondary_Male',
    'Completion_Rate_Upper_Secondary_Female',
    'Unemployment_Rate'
]

required_columns_world_data = [
    'Country',
    'GDP',
    'Population',
    'Birth Rate',
    'Co2-Emissions',
    'Life expectancy',
    'Unemployment rate',
    'Urban_population',
    'Gross primary education enrollment (%)',
    'Gross tertiary education enrollment (%)',
    'Fertility Rate',
    'Tax revenue (%)'
]

# Remove unnecessary columns from each dataset
countries_df = countries_df[required_columns_countries]
edu_df = edu_df[required_columns_education]
world_data_df = world_data_df[required_columns_world_data]
```

```python
# Display all columns from the countries dataset
print("Columns in countries dataset:")
print(countries_df.columns.tolist())

# Display all columns from the education dataset
print("\nColumns in education dataset:")
print(edu_df.columns.tolist())

# Display all columns from the world data dataset
print("\nColumns in world data dataset:")
print(world_data_df.columns.tolist())
```

```python
# Define file paths for saving the cleaned datasets
countries_file_path = "cleaned_countries_data.csv"
edu_file_path = "cleaned_education_data.csv"
world_data_file_path = "cleaned_world_data.csv"

# Save cleaned datasets to CSV files
countries_df.to_csv(countries_file_path, index=False)
edu_df.to_csv(edu_file_path, index=False)
world_data_df.to_csv(world_data_file_path, index=False)

print("Cleaned datasets saved successfully.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualization: Scatter plot of GDP per capita vs. Education expenditure (% of GDP) with logarithmic y-axis
plt.figure(figsize=(10, 6))
sns.scatterplot(data=countries_df, x='GDP per capita (current US$)', y='Education: Government expenditure (% of GDP)')
plt.title('Scatter plot of GDP per capita vs. Education expenditure')
plt.xlabel('GDP per capita (current US$)')
plt.ylabel('Education expenditure (% of GDP)')
plt.yscale('log')  # Set y-axis to logarithmic scale
plt.grid(True)
plt.savefig('scatter.svg')  # Save the plot as SVG file
plt.show()
```

```python

# Visualization : Histogram of Life expectancy distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=world_data_df, x='Life expectancy', bins=20, kde=True)
plt.title('Histogram of Life expectancy distribution')
plt.xlabel('Life expectancy')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
top_10_countries = countries_df.sort_values(by='GDP per capita (current US$)', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_10_countries, x='country', y='GDP per capita (current US$)')
plt.title('Top 10 Countries by GDP per Capita')
plt.xlabel('Country')
plt.ylabel('GDP per Capita (current US$)')
plt.xticks(rotation=45)
plt.show()
```

This line plot illustrates the trend of CO2 emissions over time. The x-axis represents the years, while the y-axis shows the CO2 emissions in million tons/tons per capita. By examining the trend of the line, we can assess whether CO2 emissions have been increasing, decreasing, or remaining relatively stable over the observed time period. Additionally, the logarithmic scale on the y-axis allows for better visualization of changes in CO2 emissions, especially when dealing with a wide range of values.

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=world_data_df, x=world_data_df.index, y='Co2-Emissions')
plt.yscale('log')  # Set y-axis scale to logarithmic
plt.title('CO2 Emissions Over Time')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions (million tons/tons per capita)')
plt.grid(True)
plt.show()
```

```python
# Create subplots
fig, axes = plt.subplots(3, 1, figsize=(10, 18))

# Plot Primary Enrollment
sns.scatterplot(data=countries_df, x='GDP per capita (current US$)', y='Education: Primary gross enrol. ratio (f/m per 100 pop.)', ax=axes[0])
axes[0].set_title('Primary Education Enrollment')
axes[0].set_xlabel('GDP per capita (current US$)')
axes[0].set_ylabel('Primary Enrollment Ratio')
axes[0].set_yscale('log')

# Plot Secondary Enrollment
sns.scatterplot(data=countries_df, x='GDP per capita (current US$)', y='Education: Secondary gross enrol. ratio (f/m per 100 pop.)', ax=axes[1])
axes[1].set_title('Secondary Education Enrollment')
axes[1].set_xlabel('GDP per capita (current US$)')
axes[1].set_ylabel('Secondary Enrollment Ratio')
axes[1].set_yscale('log')

# Plot Tertiary Enrollment
sns.scatterplot(data=countries_df, x='GDP per capita (current US$)', y='Education: Tertiary gross enrol. ratio (f/m per 100 pop.)', ax=axes[2])
axes[2].set_title('Tertiary Education Enrollment')
axes[2].set_xlabel('GDP per capita (current US$)')
axes[2].set_ylabel('Tertiary Enrollment Ratio')
axes[2].set_yscale('log')

plt.tight_layout()
plt.show()
```
